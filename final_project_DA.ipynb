## ‚öñÔ∏è Bivariate Analysis ‚Äî Player Performance Relationships

In this section, we explore how various player attributes are related to each other.  
This helps us identify patterns such as:
- Whether players with higher **Overall ratings** also tend to have better **Shooting, Passing, or Dribbling** skills.
- How **Age** relates to attributes like **Pace** or **Physical Strength**.
- Which attributes are most strongly correlated overall.

These analyses reveal how different performance metrics interact, offering insights into what contributes most to a player‚Äôs overall ability.


```python
plt.style.use('seaborn-v0_8')
sns.set_palette("Set2")

#  Overall vs Shooting
plt.figure(figsize=(7,5))
sns.scatterplot(x='overall', y='shooting', data=df, alpha=0.7)
plt.title("Overall Rating vs Shooting")
plt.xlabel("Overall Rating")
plt.ylabel("Shooting")
plt.show()
```

```python
# Overall vs Passing
plt.figure(figsize=(7,5))
sns.scatterplot(x='overall', y='passing', data=df, alpha=0.7)
plt.title("Overall Rating vs Passing")
plt.xlabel("Overall Rating")
plt.ylabel("Passing")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='pace', data=df, alpha=0.7)
plt.title("Age vs Pace (Speed)")
plt.xlabel("Age")
plt.ylabel("Pace")
plt.show()
```

```python
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Player Attributes")
plt.show()
```

```python

plt.style.use('seaborn-v0_8')
sns.set_palette("muted")

#  Group players into skill levels based on Overall rating
bins = [0, 70, 80, 90, 100]
labels = ['Average', 'Good', 'Very Good', 'Elite']
df['skill_level'] = pd.cut(df['overall'], bins=bins, labels=labels)

# Average attributes per skill level
skill_avg = df.groupby('skill_level')[['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']].mean().reset_index()

#  Bar chart comparing key attributes across skill levels
plt.figure(figsize=(10,6))
skill_avg.set_index('skill_level').plot(kind='bar', figsize=(10,6))
plt.title("Average Attributes by Player Skill Level")
plt.ylabel("Average Rating")
plt.xlabel("Skill Level")
plt.xticks(rotation=0)
plt.legend(title="Attributes")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

#  Age vs Overall (Average by age)
age_vs_overall = df.groupby('age')['overall'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(x='age', y='overall', data=age_vs_overall, marker='o')
plt.title("Average Overall Rating by Age")
plt.xlabel("Age")
plt.ylabel("Average Overall")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

#  Relationship between Shooting and Passing (averaged by skill level)
plt.figure(figsize=(8,5))
sns.barplot(x='skill_level', y='shooting', data=df, color='skyblue', label='Shooting')
sns.barplot(x='skill_level', y='passing', data=df, color='lightcoral', alpha=0.7, label='Passing')
plt.title("Shooting vs Passing across Skill Levels")
plt.xlabel("Skill Level")
plt.ylabel("Average Score")
plt.legend()
plt.show()
```

### üéØ Insight 1: Attribute Comparison by Skill Level
From the bar chart, we can observe that **Elite players** (Overall rating > 90) outperform others significantly in every key attribute ‚Äî especially *dribbling*, *passing*, and *shooting*.  
This shows that high overall ratings are strongly linked with balanced skill growth across all attributes, not just one strength.  

Meanwhile, **Average players (Overall < 70)** tend to have higher *pace* and *physical* ratings compared to their *passing* or *shooting* abilities, suggesting a difference in play style or position.


### ‚öΩ Insight 2: Average Overall Rating by Age
The line plot reveals that players generally reach their **peak overall performance between ages 25 and 30**.  
After 30, there‚Äôs a gradual decline in average overall rating ‚Äî consistent with real-world football trends where physical ability and agility start to drop, even if technical skills remain strong.  
This trend highlights how **age** plays a crucial role in performance sustainability in football.


### üèÜ Insight 3: Shooting vs Passing Across Skill Levels
When comparing *Shooting* and *Passing* across skill levels, the chart clearly shows that as players move from Average ‚Üí Elite, both metrics improve steadily.  
However, **Shooting** tends to rise more sharply than Passing, suggesting that elite players usually have better finishing and goal-scoring capabilities.  
This aligns well with real FIFA player performance data, where top attackers dominate in shooting metrics.


### üß© Insight 4: Additional Observation
From the previous analyses, we can infer that:
- A player‚Äôs **overall performance is multi-dimensional**, not dependent on a single stat.
- Attributes like *dribbling* and *passing* show the strongest correlation with higher skill levels.
- Player development peaks around age 28, where both experience and physical condition align optimally.

These findings provide actionable insight for football data analysts and game developers focusing on player rating systems.


## üîç Multivariate Analysis
Now we analyze how multiple player attributes are related to each other using correlation matrices and multivariate visuals.  
The goal is to identify which performance metrics have the strongest influence on overall rating and how they interact across player profiles.


```python
plt.style.use('seaborn-v0_8')
sns.set_palette("coolwarm")

numeric_cols = [
    'overall', 'age', 'height_cm', 'weight_kg', 
    'pace', 'shooting', 'passing', 'dribbling', 
    'defending', 'physic', 'attacking', 'skill', 
    'movement', 'power', 'mentality'
]

corr_matrix = df[numeric_cols].corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Player Attributes", fontsize=14)
plt.show()
```

```python
sns.pairplot(df[['overall', 'pace', 'shooting', 'passing', 'dribbling']], diag_kind='kde', corner=True)
plt.suptitle("Pairwise Relationships Among Key Attributes", y=1.02, fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='pace', y='dribbling', hue='overall', size='overall',
                data=df, palette='Spectral', alpha=0.7, sizes=(40, 200))
plt.title("Pace vs Dribbling Colored by Overall Rating", fontsize=13)
plt.xlabel("Pace")
plt.ylabel("Dribbling")
plt.legend(title='Overall', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### üìä Insight 5: Correlation Heatmap
From the correlation heatmap, **overall rating** shows strong positive correlation with  
`shooting`, `passing`, `dribbling`, and `pace`, while weaker correlation with `defending` and `physic`.  
This proves that offensive skills contribute the most to player ratings in FIFA.

---

### ‚öΩ Insight 6: Pairwise Analysis
The pairplot indicates that players with higher `dribbling` also tend to have better `shooting` and `passing` attributes,  
forming a clear performance cluster among elite players.

---

### üî• Insight 7: Pace vs Dribbling vs Overall
The color-coded scatter plot shows that players with both **high pace and high dribbling** almost always  
fall in the top overall ratings.  
This demonstrates how **speed and control together** define elite player quality.


## üèÜ Special Visualizations ‚Äî Top Players and Performance Trends
In this section, we focus on ranking-based and categorical visualizations.  
We analyze top-performing players, distribution across nationalities, and key rating patterns.  
These visuals help highlight real-world insights in a more engaging and presentation-friendly way.


```python
plt.style.use('seaborn-v0_8')
sns.set_palette("deep")

top10 = df.nlargest(10, 'overall')[['long_name', 'overall', 'age', 'club_name', 'nationality_name']]

plt.figure(figsize=(10,6))
sns.barplot(x='overall', y='long_name', data=top10, palette='viridis')
plt.title("Top 10 Players by Overall Rating")
plt.xlabel("Overall Rating")
plt.ylabel("Player Name")
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40]
labels = ['<20', '20-25', '25-30', '30-35', '35+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

age_group_avg = df.groupby('age_group')[['pace','shooting','passing','dribbling','defending','physic']].mean().reset_index()

plt.figure(figsize=(10,6))
age_group_avg.set_index('age_group').plot(kind='bar', figsize=(10,6))
plt.title("Average Player Attributes by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Rating")
plt.legend(title="Attributes")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
nation_stats = df.groupby('nationality_name').agg(
    player_count=('overall', 'count'),
    avg_overall=('overall', 'mean')
).reset_index()

nation_stats = nation_stats[nation_stats['player_count'] >= 10]

top_nations_balanced = nation_stats.sort_values(by='avg_overall', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='avg_overall', y='nationality_name', data=top_nations_balanced, palette='crest')
plt.title("Top 10 Nationalities by Average Overall (Min. 10 Players)")
plt.xlabel("Average Overall Rating")
plt.ylabel("Nationality")
plt.show()
```

### üèÖ Insight 8: Top 10 Players by Overall Rating
This chart highlights the highest-rated players in FIFA 22.  
We can observe that elite players not only have top overall ratings but also belong to top European clubs and are typically aged between 25‚Äì32 ‚Äî the footballing prime years.


### ‚öñÔ∏è Insight 9: Attribute Trends by Age Group
The grouped bar chart reveals that players between **25‚Äì30 years** have the best combination of all key attributes ‚Äî pace, shooting, dribbling, and physicality.  
After 30, there‚Äôs a steady decline across most metrics, especially in pace and dribbling.

### üåç Insight 10: Nationality vs Average Performance (Balanced View)
In this improved visualization, we include only nations with at least 10 players to ensure fair comparison.  
Now, major football nations like **France, Spain, Brazil, and Argentina** appear among the top,  
indicating consistent high-quality performance across multiple players rather than outliers.  
This gives a more realistic picture of player distribution and national football strength.
